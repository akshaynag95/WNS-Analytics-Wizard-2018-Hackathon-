```python
# importing required libriaries 
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
# importing train dataset
WNS_train_file = pd.read_csv('WNS_Train.csv')

# setting train file as dataframe

WNS_train_dframe = DataFrame(WNS_train_file)

WNS_train_dframe.head()
```

```python
X = WNS_train_dframe.iloc[:,[2,3,4,5,6,7,8,9,10,13,14,15,18,19,20,21,22,23,24]].values

Y = WNS_train_dframe.iloc[:,[25]].values

# splitting train dataset to check model accuracy
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.25, random_state = 0)
```

```python
SC = StandardScaler()

X_train = SC.fit_transform(X_train)
X_test = SC.fit_transform(X_test)
```

```python
RF_Classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state = 0)
```

```python
RF_Classifier.fit(X_train, Y_train)
```

```python
Y_pred = RF_Classifier.predict(X_test)

Y_pred
```

```python
WNS_train_dframe.groupby('is_promoted').mean()
```

```python
from sklearn.metrics import confusion_matrix

RF_confusion_matrix = confusion_matrix(Y_test, Y_pred)

RF_confusion_matrix
```

```python
RF_Classifier.score(X_test, Y_test)
```

```python
RF_acc = accuracy_score(Y_test, Y_pred)

RF_acc
```

```python
from sklearn.metrics import f1_score

f1_score(Y_test, Y_pred, average='micro')
```

```python
f1_score(Y_test, Y_pred, average='weighted')
```

```python
# importing test dataset
WNS_test_file = pd.read_csv('WNS_Test.csv')

# setting test file as dataframe

WNS_test_dframe = DataFrame(WNS_test_file)

WNS_test_dframe.head()
```

```python
X1 = WNS_test_dframe.iloc[:,[2,3,4,5,6,7,8,9,10,13,14,15,18,19,20,21,22,23,24]].values

X1 = SC.fit_transform(X1)
```
